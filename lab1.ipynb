## Zad. 1
Zaimplementuj funkcję bin_from_int(x, n), gdzie x to liczba całkowia nieujemna, n to liczba cyfr binarnych w wyniku. Funkcja zwraca listę cyfr binanrnych stanowiących reprezentację binanrną x. Zwracana lista ma mieć conajmniej n elementów (w razie czego uzupełniamy zerami).

bin_from_int(25, 8) -> [0, 0, 0, 1, 1, 0, 0, 1]

```python
def bin_from_int(x: int, n: int) -> list:
  binary_list = []
  while x > 0:
      binary_list.append(x % 2) # modulus
      x //= 2 # floor division

  # if the list size is less than n: add zeros to the list
  while len(binary_list) < n:
      binary_list.append(0)

  # reversing the list to get the correct order
  return binary_list[::-1]
```

```python
bin_from_int(25, 8)
```

```python
bin_from_int(71, 4)
```

```python
bin_from_int(86, 10)
```

## Zad. 2
Zaimplementuj funkcję int_from_bin(b), gdzie b to lista cyfr zapisu binarnego, a zwracany wynik to liczba odpowiedająca temu zapisowi. Oczywiście musi zachodzić x == int_from_bin(x, n).

int_from_bin([1,1,0,0,1]) -> 25

```python
def int_from_bin(b: list) -> int:
  result = 0
  for digit in b:
    result *= 2
    result += digit

  return result
```

```python
int_from_bin([1, 1, 0, 0, 1])
```

```python
int_from_bin([0, 1, 0, 0, 0, 1, 1, 1])
```

```python
int_from_bin([1, 0, 1, 0, 1, 1, 0])
```
